Mental health has become a growing concern among students in recent years, largely due to increasing academic pressure, evolving lifestyles, and various personal and social challenges. Students today are expected to maintain high academic performance while simultaneously managing expectations from family, peers, and institutions. This often leads to stress, anxiety, and, in some cases, depression, which can significantly affect their overall well-being and daily functioning. Along with these factors, lifestyle aspects such as sleep quality and habits also play a crucial role in shaping a student’s mental health.

This project aims to analyze a comprehensive student mental health dataset to better understand the relationships between key factors such as CGPA (academic performance), stress levels, anxiety, depression, and sleep quality. Furthermore, the study also examines how mental health factors interact with lifestyle indicators such as sleep quality. Overall, this project seeks to provide a holistic understanding of the interplay between academic performance and mental health. A. Handling missing values

```python
df['CGPA'] = pd.to_numeric(df['CGPA'], errors='coerce')
df['CGPA'] = df['CGPA'].fillna(df['CGPA'].mean())

print(df)
```

```python
df['Substance_Use'] = df['Substance_Use'].replace(' ', 'No Information')
df['Substance_Use'] = df['Substance_Use'].fillna('No Information')
```

```python
df.isna().sum()
```

```python
df.head(15)
```



*  In CGPA missing values is handled by giving average of CGPA
*  In substance use missing values is replaced with 'no information'.



B. Correcting data type

Sorted the table on the basis of age in assending order.

```python
df = df.sort_values(by='Age')
```

C. Creating derived columns

Creating extra 3 columns as 'total score', 'Average score'and 'mental health status'.

```python
df['Total_Score'] = df['Stress_Level'] + df['Depression_Score'] + df['Anxiety_Score']
```

```python
df['Average_Score'] = (
    df['Stress_Level'] + df['Depression_Score'] + df['Anxiety_Score']
) / 3
```

```python
df['Mental_Health_Status'] = df['Total_Score'].apply(
    lambda x: 'High Risk' if x > 10 else 'Moderate' if x > 5 else 'Low'
)
```

Students performance status

```python
conditions = [
    df['CGPA'] < 2,
    (df['CGPA'] >=2 ) & (df['CGPA'] <= 3.5),
    df['CGPA'] > 3.5
]

choices = ['Below Average', 'Average', 'Above Average']

df['Performance'] = np.select(conditions, choices, default='Unknown')
```

B. Mental Health Impact

```python
sleep_quality_map = {'Poor': 1, 'Average': 2, 'Good': 3}
df['Sleep_Quality_Num'] = df['Sleep_Quality'].map(sleep_quality_map)

plt.figure(figsize=(6,4))
sns.scatterplot(x='Stress_Level', y='Sleep_Quality_Num', hue='Course', data=df)
plt.title("Stress vs Sleep Quality")
sns.set_style("whitegrid")
plt.show()
```



*   The points are widely scattered, with no clear linear pattern.



*   High stress (4-5) is often associated with low sleep quality (1), especially visible for some courses.


*   At the same time, there are also cases where moderate stress (2-3) corresponds to good sleep quality (3).
*   There is a slight negative trend


*   As stress increases, sleep quality tends to decrease.
*   Some individuals with low stress (0-1) still have poor sleep quality (1-2).


*   Some with higher stress still report good sleep (3).
*   There is a weak negative relationship between stress and sleep quality.



*   Higher stress may reduce sleep quality, but the effect is inconsistent across individuals and courses.

C. Anxiety & Depression Trends

```python
plt.figure(figsize=(6,4))
sns.lineplot(x='Anxiety_Score', y='Sleep_Quality_Num', data=df, label='Anxiety')
sns.lineplot(x='Depression_Score', y='Sleep_Quality_Num', data=df, label='Depression')
plt.title("Mental Health vs Sleep Quality")
plt.legend()
sns.set_style("whitegrid")
plt.show()
```

Anxiety v/s sleep quality:


*   Sleep quality shows a slight increasing trend as anxiety scores raise from 0-3
*   Overall, anxiety does not strongly reduce sleep quality, the variation is minimal.

Depression v/s sleep quality:


*   Sleep quality increases initially from score 0-1
*   Then it gradually declines, with a noticeable drop at score 4.


*   This suggests that higher depression levels may negatively affect sleep quality.
*   At lower scores, both anxiety and depression show similar sleep quality level.


*   At higher scores, depression shows a clearer decline, while anxiety remains relatively stable.
